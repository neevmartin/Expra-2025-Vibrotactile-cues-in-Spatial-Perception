```python
import pandas as pd
import numpy as np
import sys
```

```python
DATA_FOLDER = "preprocessed_avoiding"
```

```python
for cond in ["ai", "ri", "ad", "rd"]:

    data = pd.read_csv(f"{DATA_FOLDER}/{cond}.csv")
    mappings = data.groupby("participant_id", as_index=False).apply(lambda df: df["natural_mapping"].iloc[0])
    mappings.columns = ['ID', 'natural_mapping']
    
    counts = mappings.groupby(["natural_mapping"]).count().T

    dir = int(counts["direct"].iloc[0])
    rev = int(counts["reversed"].iloc[0])
    sys.stdout.write(f"{cond}; Total: {dir+rev}; Direct: {dir}; Reversed: {rev}")
    
```
